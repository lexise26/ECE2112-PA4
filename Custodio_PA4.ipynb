The task involves analyzing a dataset by creating specific data frames and visualizations. The approach includes using data wrangling techniques to prepare the dataset and applying visualization techniques to explore how different features—such as track, gender, and hometown—affect average grades.

#### Expected Input:
- Dataset with columns: Name, Gender, Track, Electronics, GEAS, Average, Hometown.
- Criteria:

Instru Data Frame:
1. Students in "Instrumentation" from "Luzon" with Electronics scores >70.
2. Mindy Data Frame: Female students in "Communication" from "Mindanao" with average grades ≥55.

#### Expected Output:
- Data Frames:

1. Instru Data Frame: Contains columns Name, GEAS, Electronics >70, Track, Hometown.
2. Mindy Data Frame: Contains columns Name, Track, Electronics, Average >=55, Hometown, Gender.

- Visualizations:

1. Box Plot of Average Grades by Track: Shows grade distribution across tracks.
2. Box Plot of Average Grades by Gender: Compares average grades between genders.
3. Box Plot of Average Grades by Hometown: Displays grade distribution by hometown

```python
#Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading/reading the dataframe 

board = pd.read_excel('board2.xlsx')     #naming the dataframe as board
board                                    #displaying the dataframe
```

### 1. Create the following data frames based on the format provided:

#### Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
Instru = board[(board['Electronics'] >= 71) & (board['Track'] == 'Instrumentation') & (board['Hometown'] == 'Luzon')]
Instru.drop(columns=['Gender', 'Track', 'Hometown', 'Math', 'Communication']).reset_index(drop=True)
```

#### Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
board['Average'] = (board['Math'] + board['Electronics'] + board['GEAS'] + board['Communication'])/4
board
```

```python
Mindy = board[(board['Average'] >= 55) & (board['Hometown']=='Mindanao') & (board['Gender']=='Female')]
Mindy.drop(columns=['Gender', 'Hometown', 'Math', 'GEAS', 'Communication']).reset_index(drop=True)
```

### 2. Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
x = board['Average']

plt.figure(figsize=(20,10))
plt.style.use('fivethirtyeight')

plt.title('The Effect of Track, Gender, and Hometown to Average Score', fontsize=30, fontweight = 'bold')
plt.barh(board['Track'], x, label="Track")
plt.barh(board['Gender'], x, label="Gender")
plt.barh(board['Hometown'], x, label="Hometown")
plt.ylabel('Features', fontsize=20)
plt.xlabel('Average', fontsize=20)

plt.legend(fontsize=20)
plt.show()
```

### Based on the constructed data visualization above, it shows that there is a slight gap between the chosen college Track, student's Gender, and Hometown to the  attained Average Score. Students who chose Microelectronics track in college, a Female, and lives in Luzon are the ones who got the highest average score among the others.
